Face Detection using OpenCV

Loading the cascade classifier for face detection
Loading the images
Using cascade classifier detect the face
Display the image with rectangular box applied on it

```python
facecascade=cv2.CascadeClassifier('/content/sample_data/facecascade.xml')
```

```python
facecascade
```

```python
person=cv2.imread('/content/sample_data/personface1.jfif')
```

```python
plt.imshow(person)

```

```python
person11=cv2.cvtColor(person,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(person11)
```

```python
imagess=cv2.cvtColor(person,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagess)
```

Detect the face using cascade classifier


```python
face=facecascade.detectMultiScale(imagess,1.4,4)
```

```python
face
```

Coordinates of the rectangle are stored in the face

```python
plt.imshow(face)
```

Drawing rectangle around the face

```python
for (x,y,w,z) in face:
  cv2.rectangle(person,(x,y),(x+w,y+z),(0,0,0))
plt.imshow(person)
cv2.waitKey()
```
